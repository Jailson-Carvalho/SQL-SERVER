### 01. O que é o JOIN

O JOIN no SQL é uma forma de combinar dados de duas ou mais tabelas usando uma coluna em comum entre elas. 

Ele permite relacionar informações espalhadas no banco e devolver tudo em um único resultado. 

Na prática, o JOIN conecta tabelas diferentes como se fosse um “elo”, permitindo consultar dados completos sem duplicação ou perda de contexto.

### 06. Principais Formas de Usar o JOIN

O JOIN é uma das ferramentas mais importantes do SQL, pois permite combinar informações de tabelas diferentes a partir de um campo em comum. 

Existem várias formas de utilizá-lo, cada uma adequada a um tipo de necessidade de análise.

O INNER JOIN é o mais comum e retorna apenas os registros que existem simultaneamente nas duas tabelas. 

Já o LEFT JOIN retorna todos os registros da tabela da esquerda, mesmo que não haja correspondência na tabela da direita — útil para identificar dados faltantes ou cruzar bases principais com complementares. 

O RIGHT JOIN funciona de forma inversa, trazendo todos os registros da tabela da direita. 

O FULL JOIN reúne tudo: registros que coincidem e também os que não coincidem em nenhuma das tabelas. 

Além disso, existem o CROSS JOIN, que combina todos os registros de uma tabela com todos os da outra, e o SELF JOIN, onde a mesma tabela é unida a ela mesma para comparar registros internos. 

Cada tipo de JOIN serve a um cenário diferente, ajudando a responder perguntas específicas e oferecendo flexibilidade total na análise de dados.
