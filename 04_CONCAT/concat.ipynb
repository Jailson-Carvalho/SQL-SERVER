# CONCAT

### 1. introdução ao CONCAT

A função CONCAT() no SQL Server é uma ferramenta essencial para transformar dados brutos em informações mais claras e personalizadas. 

Ela permite unir textos, espaços e valores de diferentes colunas em uma única string, criando frases completas e campos derivados que facilitam a compreensão dos dados. 

Com o CONCAT(), é possível montar nomes completos, construir mensagens explicativas e apresentar informações de forma mais amigável e profissional, tornando a consulta muito mais informativa e útil em relatórios e análises.

### 4. CONCAT – Construção de Textos Personalizados


Este é o bloco mais importante do código, onde você utiliza a função CONCAT(), que serve para juntar textos e formar frases completas diretamente no resultado da consulta.

O que é CONCAT?

A função CONCAT() recebe dois ou mais valores e os une em uma única string.
Ela é extremamente útil para criar textos explicativos, gerar campos derivados ou personalizar informações para relatórios.

### 4.1. Criando o Nome Completo

            CONCAT(FirstName, ' ', LastName) AS 'Nome Completo'

Aqui você junta o nome e o sobrenome, adicionando um espaço entre eles.

O resultado é um campo informativo e de fácil leitura.

### 4.2. Criando um Texto Explicativo sobre E-mail

            CONCAT('O e-mail de ', ' ', FirstName, ' ', 'é:', ' ', EmailAddress) AS 'E-mail'
Essa linha monta uma frase completa, como:

“O e-mail de Maria é: maria.silva@contoso.com

Perceba como o CONCAT() permite combinar:

            textos fixos

            espaços

            dados das colunas

Tudo isso formando uma mensagem clara e amigável.

### 4.3. Construindo uma Frase com o Endereço

            CONCAT('O endereço de ', ' ', FirstName, ' ', 'é: ', AddressLine1) AS 'Endereço'

Da mesma forma, esta instrução cria um texto descritivo para o endereço do cliente.
É especialmente útil para relatórios personalizados ou integrações que exigem textos completos.
