# VIEW

Uma VIEW é uma tabela virtual criada a partir de uma consulta (SELECT), que não armazena dados por conta própria, mas sempre mostra informações atualizadas da(s) tabela(s) original(is).

Para que serve:

    - Facilita consultas complexas

    - Organiza e padroniza cálculos

    - Melhora a segurança (mostra só o que o usuário precisa ver)

    - Evita repetir SELECTs grandes no sistema

### 01 View SQL Server para o Power BI

Após a nossa query pronta, criaremos a VIEW  e vamos carregar ela no SQL Server.

Vantangens dessa ação:
    Melhor desempenho no Power BI: o Power BI consulta a VIEW como se fosse uma tabela limpa e já tratada;

    Reduz a complexidade: qualquer cálculo, JOIN ou conversão já fica resolvido dentro da VIEW;

    Evita repetir lógica: Se vários relatórios usam o mesmo cálculo (ex.: horas, tempo, produtividade), a VIEW garante que todos trabalhem com a mesma regra;

    Segurança e controle: você pode permitir que o Power BI enxergue apenas a VIEW, sem liberar acesso às tabelas completas.


vamos fazer primeiro a criação da View e logo após carrewgaremos no Power BI.

### 03.1 Sobre a Query

Modelagem de Dados para Indicadores de Produção

✅ Explicação Completa do Fluxo Lógico da Query

Essa query foi construída com o objetivo de gerar uma tabela fato de produção, consolidando informações operacionais importantes para análise, como operador, recurso, quantidade produzida, ocorrências e tempo total de produção.
A seguir, o fluxo lógico do que a query faz:

✅ 1. Seleção do Banco de Dados

A instrução USE Producao define que todas as operações seguintes serão realizadas dentro do banco Producao, garantindo que a consulta seja executada no ambiente correto.

✅ 2. Seleção dos Campos da Tabela Fato

A query utiliza a tabela OrdemProducaoLancamento (T1) como base.

 Dela são extraídas informações operacionais como:

        ID_OP – código da ordem de produção
        ID_Operador – responsável pela operação
        ID_Recurso – máquina ou posto produtivo
        ID_Ocorrencia – tipo de ocorrência registrada
        QtdeProduzida e QtdeRefugada – desempenho e perdas
        Datas e horários de início e fim da operação
        Esses campos compõem a estrutura principal da tabela fato.

✅ 3. Enriquecimento dos Dados via JOIN

A query realiza duas junções para enriquecer os dados:

        🔹 JOIN com OrdemProducao (T2)
        Traz o ID_Produto associado à ordem — informação essencial para análises por item, família ou linha de produto.
        🔹 JOIN com Ocorrencia (T3)
        Traz a descrição textual da ocorrência, permitindo análises mais intuitivas (ex.: parada técnica, retrabalho, setup, operação normal).
        Esses JOINs transformam o lançamento bruto em um dataset analítico completo.

✅ 4. Tratamento das Datas e Horas

As colunas DataHoraInicial e DataHoraFinal são quebradas em:
        Data Inicial
        Hora Inicial
        Data Final
        Hora Final
Essa separação facilita visualizações em dashboards e cálculos de produtividade.

✅ 5. Cálculo do Tempo Total da Operação (HH:MM)

A parte mais técnica da query é justamente o cálculo do tempo:
CONVERT(CHAR(5),
DATEADD(SECOND, DATEDIFF(SECOND, T1.DataHoraInicial, T1.DataHoraFinal), '00:00:00'), 108 use) AS 'Horas'

Esse código:
        Calcula a diferença entre início e fim em segundos
        Converte esses segundos em formato de horário
        Formata o resultado final em HH:MM
        Assim, tempos como 50 minutos ou 1h30 são exibidos como:
        00:50
        01:30

Perfeito para relatórios de produtividade.

✅ Resumo Geral

Essa query constrói uma base analítica robusta que:

        ✅ Consolida dados de produção
        ✅ Enriquece com dados de produto e ocorrência
        ✅ Padroniza datas e horas
        ✅ Calcula o tempo exato de cada operação
        ✅ Prepara tudo para análises em SQL, Power BI ou dashboards
        
É um modelo ideal para tabelas fato de produção em projetos de BI ou otimização industrial.
