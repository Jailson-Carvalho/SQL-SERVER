## 1: Ordem de Produção (dbo.OrdemProducaoLancamento):
Controla cada lote ou pedido de fabricação, com o produto, quantidade total a produzir e datas previstas.



```python
from IPython.display import Image, display

caminho_imagem = r"C:\Users\Jailson\Downloads\Formação Analista de Dados\03 Módulo SQL Banco de Dados\querys\Slide1.JPG"
display(Image(filename=caminho_imagem))

```

#### Descrição das colunas da tabela:

- ID_OP:	Identificador da Ordem de Produção — cada número representa uma ordem específica criada para produzir um determinado item.
- ID_Operador:	Código do operador (funcionário) responsável pela execução da operação de produção.
- ID_Ocorrencia:	Código de ocorrência registrada durante o processo (como paradas, falhas, manutenções ou eventos de produção). Pode ser NULL se não houve ocorrência.
- ID_Recurso:	Identificador do recurso utilizado (ex.: máquina, estação de trabalho ou equipamento). Indica onde a operação foi realizada.
- DataHoraInicial:	Data e hora de início da execução da operação na ordem de produção. Mostra o momento em que o operador começou o processo.
- DataHoraFinal:	Data e hora de término da execução. Permite calcular a duração da operação e o tempo total de produção.
- QtdeProduzida:	Quantidade de itens produzidos com sucesso nessa execução (sem defeito).
- QtdeRefugada:	Quantidade de itens refugados ou rejeitados — peças que apresentaram defeito e não foram aprovadas na inspeção.

#### Descrição das colunas da tabela:
- ID_Produto:	Código único que identifica o produto no sistema. Pode representar um item acabado, componente intermediário ou matéria-prima. É utilizado como chave primária e pode ser referenciado em tabelas como OrdemProducao.
- Descricao:	Nome ou descrição detalhada do produto. Contém informações úteis para identificação, como modelo, tipo de material, processo de fabricação (ex.: “CORPO DE PINÇA”, “GANCHEIRA ALTA”).
- Pecas_Hora:	Quantidade de peças que podem ser produzidas por hora, conforme o planejamento de produção. Esse valor serve como indicador de capacidade produtiva (ritmo padrão de produção) e pode ser usado para medir eficiência real versus teórica.
