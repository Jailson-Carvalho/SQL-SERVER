### 2. CREATE TABLE (Criação da tabela)

O comando CREATE TABLE é usado no SQL para criar uma tabela, definindo suas colunas, tipos de dados e regras.
 
Ele é essencial porque estabelece a estrutura onde as informações serão armazenadas. 

Uma tabela bem criada garante organização, consistência e facilita consultas, análises e integrações dentro do banco de dados.

#### Explicando o código para a criação da tabela:


			USE teste (Aqui eu disse para o SQL que quero usar o Banco de Dados: teste. Que acabei de criar)

			-- CREATE TABLE (Isso é um comentário, o SQL não vai rodar, isso é apenas para nós nos lembrarmos do que estamos fazendo)

			CREATE TABLE Vendas (Traduzindo: CRIAR TABELA Vendas. em seguida abri () para escrever as colunas da tabela.)
			(
				id INT PRIMARY KEY NOT NULL (Essa coluna se chama: id. Será o id das vendas. Também disse que é uma chave primáris 'PRIMARY KEY' e não aceita informações nulas ' NOT NULL'.)
				,product_id INT NOT NULL (Coluna product_id, o tipo de dados a ser recebido é inteiro 'INT' e não permite nulidade 'NOT NULL'.)
				,Cliente VARCHAR (50) (Coluna Cliente recebe valor do tipo textocom no máximo 50 caracteres 'VARCHAR (50)')
				,data_compra DATE NOT NULL (Nessa coluna teremos a data em que o cliente foi cadastrado, é do tipo data 'DATE' e não acita valores nulos.)
				,valor_total MONEY (O valor da compra do cliente será disposto na coluna valor_total que é do tipo money 'MONEY')
			)

### 3. Inserir valores na tabela criada (insert into)

O comando INSERT INTO é usado no SQL para inserir novos registros dentro de uma tabela. 

Ele é importante porque adiciona os dados que serão armazenados, consultados e analisados pelo sistema. 

Sem o INSERT, a tabela teria apenas estrutura, mas nenhum conteúdo útil para gerar informações, relatórios ou decisões.

Na imagem abaixo fiz um SELECT * FROM vendas. Isto é: selecionei toda a tabela para verificar quais dados temos nela, e não é surpresa que a tabela está vazia, não há dados. Por isso vamos inserir dados nela.

```python
from IPython.display import Image, display

caminho_imagem = r"C:\Users\Jailson\Downloads\CONHECENDO AS TABELAS\05_CREATE_TABLE_INSERT_INTO\imagens\INSART INTO VENDAS\Slide1.JPG"
display(Image(filename=caminho_imagem))
```
